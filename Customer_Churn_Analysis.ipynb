```python
import pandas as pd

#load the dataset
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#set the customerID to index
df.set_index('customerID',inplace=True)
```

```python
#Shape of the dataset
df.shape
```

```python
df.info()
```

ensure data is consistent and free from errors

```python
#Replace empty strings with NaN
df['TotalCharges']=df['TotalCharges'].replace('',pd.NA)
```

```python
#convert to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(subset='TotalCharges',axis=0,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Check duplicate entries
df.index.duplicated().sum()
```

Convert the Churn into numeric

```python
#Encode churn: No=0, Yes=1
df['Churn'] = df['Churn'].str.strip().map({'No':0,'Yes':1})
```
